```python
# import libraries
import torch
import matplotlib.pyplot as plt
```

## 1. Create the Data

```python
# variable to evaluate over
x = torch.linspace(-3,3,101)

# create a function that returns the activated output
def NNoutputx(actfun ,arr):
    
    # get activation function type
    # this code replaces torch.relu with torch.<actfun>
    actfun = getattr(torch,actfun)
    
    return actfun( arr )
```

```python
# the activation functions
activation_funs = [ 'relu', 'sigmoid', 'tanh' ]

fig = plt.figure(figsize=(6,4))

for actfun in activation_funs:
    plt.plot(x,NNoutputx(actfun ,arr =x),label=actfun,linewidth=1.5)

# add reference lines
dashlinecol = [.7,.7,.7]
plt.plot(x[[0,-1]],[0,0],'--',color=dashlinecol)
plt.plot(x[[0,-1]],[1,1],'--',color=dashlinecol)
plt.plot([0,0],[-1,3],'--',color=dashlinecol)

# make the plot look nicer
plt.legend()
plt.xlabel('x' ,fontweight = "bold")
plt.ylabel('$\sigma(x)$' )
plt.title('Various Activation functions' ,fontweight = "bold")
#plt.xlim(x[[0,-1]])
#plt.ylim([-1,3])

plt.show()
```

## 2. More Activation Functions

```python
# create a function that returns the Activation output functions
# this is different from the previous function
def NNoutput(actfun):
    
    # this code replaces torch.nn.relu with torch.nn.<actfun>
    actfun = getattr(torch.nn,actfun)
    
    # Returning an Activation Function
    return actfun()
```

```python
# the activation functions
activation_funs = [ 'ReLU6', 'Hardshrink', 'LeakyReLU' ]

fig = plt.figure(figsize=(7,4))

for actfun in activation_funs:
    
    plt.plot(x,NNoutput(actfun)(x), label=actfun, linewidth=1.5)

# add reference lines
dashlinecol = [.7,.7,.7]
plt.plot(x[[0,-1]],[0,0],'--',color=dashlinecol)
plt.plot(x[[0,-1]],[1,1],'--',color=dashlinecol)
plt.plot([0,0],[-1,3],'--',color=dashlinecol)

# make the plot look nicer
plt.legend()
plt.xlabel('x' , fontweight = "bold")
plt.ylabel('$\sigma(x)$')
plt.title('Various Activation Functions')
# plt.xlim(x[[0,-1]])
# plt.ylim([-1,3])
plt.ylim([-.1,.1])

plt.show()
```

## 3. Difference between torch and torch.nn

```python
# redefine x (fewer points to facilitate visualization)
x = torch.linspace(-3,3,21)

# in torch
y1 = torch.relu(x)

# in torch.nn
f = torch.nn.ReLU()
y2 = f(x)

plt.figure(figsize = (5,3) , dpi = 140)
# the results are the same
plt.plot(x,y1,'ro',label='torch.relu')
plt.plot(x,y2,'bx',label='torch.nn.ReLU')
plt.legend()s
plt.xlabel('Input')
plt.ylabel('Output')

plt.show()
```
