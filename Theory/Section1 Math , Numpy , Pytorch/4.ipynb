## 13. Derivatives

```python
# Create symbolic variables in sympy
x = sym.symbols('x')

# Create a function
fx = 2*x**2

# Compute its derivative
dfx = sym.diff(fx , x)

# Print
print("Function f(x) : ",fx)
print("Differentition of f(x) :",dfx)
```

```python
# Plotting

symplot(fx , (x,-4,4),title = "Function f(x) : 2*x**2 " ,size = (3,3))

symplot(dfx,(x,-4,4) , title = "derivative of f(x)" ,size = (3,3))

plt.show()
```

```python
# Create ReLU and Sigmoid function

ReLU = sym.Max(0,x)
sigmoid =  1 /(1 + sym.exp(-x))
```

```python
# Plotting  title = "Sigmoid function"

p = symplot(ReLU,(x,-4,4) ,label = "ReLU" , size = (3,3) ,
            show = False , line_color = 'blue')

p.extend(symplot(sigmoid , (x,-4,4)  , label = "Sigmoid" ,
          show = False ,line_color = 'red' ))

p.legend = True
p.title = "The functions"
p.show()
```

#### Another way

```python
p1 = symplot(ReLU,(x,-4,4) ,label = "ReLU" , size = (3,3) ,
             show = False , line_color = 'blue')

p2 = symplot(sigmoid , (x,-4,4)  , label = "Sigmoid" , 
             show = False ,line_color = 'red' )
         
p1.extend(p2)

p1.legend =True
p1.title = "The functions"
p1.show()
```

```python
dReLU = sym.diff(ReLU ,x)
dsigm = sym.diff(sigmoid , x)
```

```python
p1 = symplot(dReLU,(x,-4,4) ,label = "d(ReLU)" , size = (3,3) ,
             show = False , line_color = 'blue')

p2 = symplot(dsigm , (x,-4,4)  , label = "d(Sigmoid)" , 
             show = False ,line_color = 'red' )
         
p1.extend(p2)

p1.legend =True
p1.title = "The Derivatives"
p1.show()
```

#### Example 1

```python
# create symbolic variable in sympy
x = sym.symbols('x')

# Create two functions
fx = 2*x**2
gx = 4*x**3 -3*x**4

# compute their individual derivatives
dfx = sym.diff(fx,x)
dgx = sym.diff(gx,x)

# Applying product rule via sympy
diff_sympy = sym.diff(fx*gx,x)

# displaying
print("The functions :")
display(fx)
display(gx)

print("differentiation of f(x).g(x) w.r.t x : ")
display(diff_sympy)
```
